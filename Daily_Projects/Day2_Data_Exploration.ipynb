```python
from sklearn.datasets import load_iris
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = load_iris()
iris
```

I will convert data to Pandas DataFrame, so i can make different data analysis with it.

```python
df = pd.DataFrame(data = iris.data, columns = iris.feature_names)
df['target'] = iris.target_names[iris.target]
df
```

```python
#here i wanted to know the difference between loc and iloc,so basically loc will return rows with the index label from 0 to 3 which are 0,1,2,3(4 rows)
df.loc[0:3]
```

```python
#here iloc will return rows with the positions in range 0 to 3 which are 0,1,2(3 rows)
df.iloc[0:3]
```

# Data Visualization

Now to the most interesting part(for me personally). Data Visualization!! I will start by easy bar chart and will go to more complex analyses. I will use seaborn and matplotlib libraries.

```python
sns.countplot(x='target', hue='target', data=df, palette='Set1')
plt.show()
```

Here i will make comparisons of pair of features based on their species(target). First between petal length and petal width,then sepal length with sepal width.

```python
sns.scatterplot(x='petal length (cm)', y='petal width (cm)', hue='target', data=df)
```

Observation between petal length and petal width i can say that:


1.   Iris Setosa's petals much smaller
2.   Iris Versicolor has meduim petal size
3.   From the chart clearly shown that Virginica has bigger petals



```python
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', hue='target', data=df)
```

Observation based on comparison between sepal length and sepal width:


1.   Even though Iris Setosa has smaller sepal length, in comparison to others it has much larger sepal width
2.   Iris Virginica's both sepal length and sepal width has big range of size

3.   Versicolor lies in the middle, meaning it has medium size on length and width



Next chart will be analysis of comaprison of all columns.

```python
sns.pairplot( hue='target', data=df, height=2) #height of each subplot 2 inches(5cm)
```

This plot shows many relationships between features and with this it is easier to understand the plot.

Next will be histograms, i can check the distrbution of data for various columns.

```python
fig, axes = plt.subplots(2,2, figsize=(10,10))
axes[0,0].set_title('Sepal Length (cm)')
axes[0,0].hist(df['sepal length (cm)'], bins=7)
axes[0,1].set_title('Sepal Width (cm)')
axes[0,1].hist(df['sepal width (cm)'], bins=7)
axes[1,0].set_title('Petal Length (cm)')
axes[1,0].hist(df['petal length (cm)'], bins=7)
axes[1,1].set_title('Petal width (cm)')
axes[1,1].hist(df['petal width (cm)'], bins=7)
```

So this plot gives information:


1.   Most of the irises(30-35) has sepal length between 5.5 - 6.0
2.   Almost 60 irises has sepal width 2.7 - 3.0 cm
3.   In 3rd plot it seen that most irises have smaller petal length size, from 1-2 cm
4.   And lastly,petal width in nearly 50 irises has small length of 0.5 cm



Distribution plot! The one of the most valuable plots, it will give information about spread,central tendency also i can assess presence of outliers and erros.

```python
plot = sns.FacetGrid(df, hue='target')
plot.map(sns.histplot, "sepal length (cm)", kde=True).add_legend()

plot1 = sns.FacetGrid(df, hue='target')
plot1.map(sns.histplot, "sepal width (cm)", kde=True).add_legend()

plot2 = sns.FacetGrid(df, hue='target')
plot2.map(sns.histplot, "petal length (cm)", kde=True).add_legend()

plot3 = sns.FacetGrid(df, hue='target')
plot3.map(sns.histplot, "petal width (cm)", kde=True).add_legend()

```

Findings from these plots:


1.   Sepal Length has huge overlapping
2.   Sepal Width has huge overlapping
3.   Petal length has little amount of overlapping
4.   Petal Width has littke amount of overlapping

Lastly,Correlation!!

```python
sns.heatmap(df.corr(method='pearson'),annot=True)
```

from Correlation i got:


1.   Sepal length has strong positive linear relationship with both petal length and petal width. That indicates, If sepal length increases, these 2 will also increase, same thing when it decreases.
2.   But for sepal width it is not the same, both petal length and petal width have moderate negative linear relationship with sepal width. If sepal width decreases these 2 will increase, and vice-versa.
